```python
from databricks.sdk import WorkspaceClient
import yaml
import os
from pathlib import Path


def init_bundle(catalog: str):
    path = Path(os.getcwd()).parent / "bundle/databricks.yml"
    data = yaml.safe_load(open(str(path)).read())
    data["targets"]["dev"]["workspace"]["host"] = WorkspaceClient().config.host
    data["targets"]["dev"]["workspace"]["catalog"] = catalog
    open(path, "w").write(yaml.safe_dump(data, sort_keys=False))


def cleanup_catalog(catalog: str):
    spark.sql(f"CREATE CATALOG IF NOT EXISTS `{catalog}`")
    spark.sql(f"DROP SCHEMA IF EXISTS `{catalog}`.bronze CASCADE")
    spark.sql(f"DROP SCHEMA IF EXISTS `{catalog}`.silver CASCADE")
    spark.sql(f"DROP SCHEMA IF EXISTS `{catalog}`.gold CASCADE")


def init_catalog(catalog: str):
    spark.sql(f"CREATE CATALOG IF NOT EXISTS `{catalog}`")
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS `{catalog}`.bronze")
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS `{catalog}`.silver")
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS `{catalog}`.gold")
```

```python
cleanup_catalog(catalog)
init_catalog(catalog)
init_bundle(catalog)
print(f"âœ… Initialized the bronze and silver bundle for catalog `{catalog}`!")
```
